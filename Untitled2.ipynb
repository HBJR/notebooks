```python
member_df = pd.read_csv('../data/H105_members.csv')
```

```python
def f(state, district):
    return '.'.join([state,str(district)])

member_df['statecd'] = member_df.apply(lambda x: f(x.state_abbrev, x.district_code), axis=1)
```

```python
df_104 = pd.read_csv('../data/H104_members.csv')

full_member_df = df_104.merge(member_df, on = ['state_abbrev', 'district_code'])
```

```python
full_member_df['nominate_1_change'] = full_member_df['nominate_dim1_y'] - full_member_df['nominate_dim1_x']
full_member_df['nominate_2_change'] = full_member_df['nominate_dim2_y'] - full_member_df['nominate_dim2_x']
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
def f(x):
    if x>0: return 1
    else: return 0

final_df['fox_availible'] = final_df['subrf1998'].map(f)
```

```python
final_df['dem_1998'] = final_df['party_code_y'].map(lambda x: 1 if x == 100 else 0)
final_df['dem_1996'] = final_df['party_code_x'].map(lambda x: 1 if x == 100 else 0)
```

```python
def f(party_1998, party_1996):
    if party_1998 == party_1996: return True
    else: return False

final_df['same_party'] = final_df.apply(lambda x: f(x.dem_1998, x.dem_1996), axis=1)
```

```python
same_df = final_df[final_df['same_party']]
```
